```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
HAHA = 300
hightcut = 1
lowcut = -0.5
```

```python
# 参数值
a = -0.000016
c = hightcut
b = (lowcut-c-HAHA*HAHA*a)/HAHA
b = -0.0002
```

```python
def f(x):
    return a * x**2 + b * x + c

# 定义x的范围
x = np.linspace(0, HAHA, 1000)

# 计算对应的y值
y = f(x)

# 绘制函数图像
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y = f(x)")
plt.title("Quadratic Function: f(x) = ax^2 + bx + c")
plt.grid()

# 标注关键点
plt.scatter([0, HAHA], [hightcut, lowcut], color="red", zorder=5)
plt.text(0, hightcut, "y(0) = "+str(hightcut), fontsize=12, verticalalignment="bottom", horizontalalignment="left")
plt.text(HAHA, lowcut, "y("+str(HAHA)+") = "+str(lowcut), fontsize=12, verticalalignment="top", horizontalalignment="right")

plt.show()
```
